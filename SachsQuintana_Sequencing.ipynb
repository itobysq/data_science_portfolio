 Since all of the questions in Part 1 relate to the aggregate of the 96 samples, the first thing that I want to do is aggregate all of the samples into one Python dataframe.

```python
import pandas as pd
import os, glob
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import scipy.stats
matplotlib.style.use('ggplot')
```

```python
path_to_files = 'C:\\Users\\toby\\Documents\\Jobsearch\\DataScience\\Counsyl\\DataAssessment\\'
os.chdir(path_to_files)
file_list = glob.glob('SEQ*.txt')
def consolidate_files(path_to_files):
    dataframes = []
    for target_file in file_list:
        full_path = path_to_files + target_file
        file_dataframe = pd.read_table(full_path)
        file_dataframe['filename'] = pd.Series([target_file for number 
                                               in range(len(file_dataframe))],
                                               index=file_dataframe.index)
        dataframes.append(file_dataframe)
    big_dataframe = pd.concat(dataframes)
    return big_dataframe
big_dataframe = consolidate_files(path_to_files)
big_dataframe = big_dataframe.reset_index(drop=True)
```

For part B, we are comparing the relationship between two categorical variables without an expectation for what that value is. To measure significance, we'll use a Fisher exact test.

```python
number_bloods = big_dataframe['sample_type'].value_counts()[0]
number_saliva = big_dataframe['sample_type'].value_counts()[1]
```

```python
# a function to generate 2x2 categorical tests
def categorical(qc_test):
    criteria1 = big_dataframe['sample_type']=='blood'
    criteria2 = big_dataframe[qc_test] == 'passed'
    criteria3 = big_dataframe['sample_type']=='saliva'
    blood_passed = float((criteria1 & criteria2).value_counts()[1])
    saliva_passed = float((criteria3 & criteria2).value_counts()[1])
    return [[blood_passed, saliva_passed],
                [number_bloods-blood_passed, number_saliva - saliva_passed],
                [(number_bloods-blood_passed)/number_bloods, (number_saliva - saliva_passed)/number_saliva]]
coverage_table = categorical('coverage')
table_df = pd.DataFrame(coverage_table, columns = ['blood sample', 'saliva sample'], 
             index= ['pass coverage', 'fail coverage', 'fraction failed'])
np.round(table_df,2)
```

```python
oddsratio, pvalue = scipy.stats.fisher_exact([coverage_table[0], coverage_table[1]])
print pvalue
```

The blood samples have a significantly higher rate of passing the coverage QC. Now to test if the sample type relates to passing coverage but failing CNV calling QC

```python
# a function to plot the failure rate of QC metrics
def plot_const_bin_width(qc_param, num_bins, second_part, hue, legend_loc):
    
    bins = np.array(np.linspace(big_dataframe[qc_param].min(),big_dataframe[qc_param].max(),num_bins))
    df_sort_frag_size = big_dataframe.sort_index(by=qc_param, ascending = True);
    df_sort_frag_size= df_sort_frag_size.reset_index(drop=True)
    frag_list =  np.array(df_sort_frag_size[qc_param].tolist())

    x_axis = []
    rate = []
    bin_size_labels = []

    for bin_index in range(0, len(bins)-1):
        slice_indices = []
        #match the bin delimiters with the frag list and determine where to slice the dataframe
        slice_indices =sorted(list(set(np.where(frag_list<bins[bin_index+1])[0].tolist()) & 
                        set(np.where(frag_list>bins[bin_index])[0].tolist())))
        if len(slice_indices)>1:
            slice_start = int(slice_indices[0])
            slice_end = int(slice_indices[-1])
            if slice_start != slice_end:   
                #obtain pass / fail values for the critieria in part 1 and part 2 of the problem
                cut_df = df_sort_frag_size[slice_start:slice_end]
                criteria1 = cut_df['coverage']=='failed'
                part1_criteria = criteria1.value_counts()
                
                criteria2 = cut_df['coverage'] == 'passed'
                criteria3 = cut_df['CNV_calling'] == 'failed'
                part2_criteria = (criteria2 & criteria3).value_counts()
                
                if second_part == 'yes':
                    pass_fail_count = part2_criteria
                    legend_text = 'pass coverage fail CNV rate'
                else:
                    pass_fail_count = part1_criteria
                    legend_text = 'coverage fail rate'

        # condition to see if all passed or all failed
                if len(pass_fail_count) == 1:
                    if str(who_passed.index[0]) == 'passed':
                        rate.append(0)
                    else:
                        rate.append(100)
                else:
                    rate.append(100*float(pass_fail_count[1])/
                                (float(pass_fail_count[0])+float(pass_fail_count[1])))
                x_axis.append(np.mean([bins[bin_index], bins[bin_index+1]]))
                bin_size_labels.append(len(cut_df))
     # now plot the data           
    fig = plt.figure(figsize=(15,5))
    ax = fig.add_subplot(111)
    ax.set_xlabel(qc_param)
    ax.set_ylabel('failure rate (%)')
    ax.plot(x_axis, rate, marker='o', color=hue,)
    matplotlib.rcParams.update({'font.size': 14})
    ax.legend([legend_text],loc=legend_loc)
    plt.title(qc_param+' failure rate, labels denote # of samples in bin, '+str(num_bins)+' bins used\n')
    
    for i, txt in enumerate(bin_size_labels):
        ax.annotate(txt, (x_axis[i],rate[i]))
    plt.show()
```

Part D: Have the metrics been stable over time? To find this, let's find the mean and stdev for each QC in each datafile and plot it

```python
data_over_time = {
                'filenames':[],
                'frag_size_means':[],
                'frag_size_95ci':[],
                'pre_quant_means':[],
                'pre_quant_95ci':[],
                'post_quant_means':[],
                'post_quant_95ci':[]
                }
files = big_dataframe['filename'].unique().tolist()
for a_file in files:
    a_file_df = pd.DataFrame(big_dataframe.loc[big_dataframe['filename']==a_file])
    data_over_time['filenames'].append(float(a_file[-7:-4]))
    data_over_time['pre_quant_means'].append(a_file_df['pre_quant'].mean())
    data_over_time['pre_quant_95ci'].append(2*a_file_df['pre_quant'].std())
    data_over_time['post_quant_means'].append(a_file_df['post_quant'].mean())
    data_over_time['post_quant_95ci'].append(2*a_file_df['post_quant'].std())
    data_over_time['frag_size_means'].append(a_file_df['frag_size'].mean())
    data_over_time['frag_size_95ci'].append(2*a_file_df['frag_size'].std())
```

```python
fig, axs = plt.subplots(nrows=3, ncols=1, sharex=True)
plt.tight_layout()

ax = axs[0]
ax.errorbar(data_over_time['filenames'],data_over_time['frag_size_means'], 
            yerr = data_over_time['frag_size_95ci'],fmt = 'o', markersize = 1)
ax.set_ylabel('frag size')

#ax.set_xlabel('file suffix')


ax = axs[1]
ax.errorbar(data_over_time['filenames'],data_over_time['pre_quant_means'], ecolor='b',
            yerr = data_over_time['pre_quant_95ci'],fmt = 'o', markersize = 1)
#ax.set_xlabel('file suffix')
ax.set_ylabel('pre quant')

ax = axs[2]
ax.errorbar(data_over_time['filenames'],data_over_time['post_quant_means'], ecolor='g',
            yerr = data_over_time['post_quant_95ci'],fmt = 'o', markersize = 1)
ax.set_xlabel('file suffix')
ax.set_ylabel('post quant')
fig.suptitle('in-process QC means with 95% confidence intervals')
```

###### Answer 1D:

###### Part 1D: To a 95% confidence interval, the QC metrics have been stable over time. However, there is an increase in the frag size centered around the 100th file that I would ask about.

### A high throughput protocol for measuring GC-content bias in samples would be to obtain a control segment of DNA that is similar to the segment of DNA that we are interested in. For example, if we were testing the DNA of a fetus to see if the baby will have a chromosome condition like down syndrome, then it would be helpful to have the same chromosome of the mother sequenced using an array or cytogenetics or any other "old-school" method of sequencing. That way, we could see if there was GC bias because we would have a reference to compare the fetus chromosome to. Then we could subtract the reads between the mother chromosome and the fetus chromosome in the areas that we are certain are not changing from mother to child to measure GC bias.

Histograms of qc metrics. I first answered 1c this way; however, it is hard to visualize a failure rate by comparing histograms

```python
def calc_bins(col_name,num_bins):
    bins =  np.linspace(big_dataframe[col_name].min(),big_dataframe[col_name].max(),num_bins)
    return bins
num_bins = 100
frag_bins = calc_bins('frag_size', num_bins)
post_quant_bins = calc_bins('post_quant', num_bins)
pre_quant_bins = calc_bins('pre_quant', num_bins)

pass_coverage_df = big_dataframe.loc[big_dataframe['coverage']=='passed']
fail_coverage_df = big_dataframe.loc[big_dataframe['coverage']=='failed']

# a function to make 1x2 histograms to compare which QC metrics are indicative of failing  / passing
#based on some condition
def histograms(in_process_qc,bins,fail_df, pass_legend, fail_legend):
    fig, axes = plt.subplots(nrows=1, ncols=2)
    ax0, ax1 = axes.flat

    ax0.hist(fail_df[in_process_qc].tolist(),bins.tolist(), alpha=0.5, label=pass_legend)
    ax0.legend(loc='upper right')
    ax0.set_ylabel('frequency')
    ax0.set_xlabel(in_process_qc)
    ylim0 = ax0.get_ylim()
    xlim0 = ax0.get_xlim()

    ax1.hist(pass_coverage_df[in_process_qc].tolist(),bins.tolist(), alpha=0.5, label=fail_legend, color='blue')
    ax1.legend(loc='upper right')
    ax1.set_xlabel(in_process_qc)
    ax1.axis ([xlim0[0],xlim0[1],ylim0[0],ylim0[1]*100])
    plt.show()

histograms('frag_size',frag_bins, fail_coverage_df, 'failed coverage', 'passed coverage')
```
