Notice that the device name follows a very specific format. That is, date, cathode, electrolyte, anode. Sometimes, the person making the device would not follow the convention perfectly. To account for this, I had to create a dictionary to help correct for errors. 

```python
dnames = df['Device Name'].tolist()
anode = [] 
electrolyte = []
cathode = []
dates = []
film_materials = {
    'cathode':['p3ht','pedot', 'pedot-A', 'pedot-N', 'pedot123dw',
              u'pedot477dw','pedot7','pedot7FS','pedot7FSmayer',
              'pedot7dw123','pedot7dw99','pedot7f','pedot7li',
              'pedot7mayer', 'pedot7pcbm','pedot9','pedot99dw',
              'pedotD','pedotF','pedotFS','pedotFS34','pedotFSmayer',
              'pedotG','pedotLi','pedotN','pedotdw477','pedotf',
              'pedotfp','pedothpcbm','pedotli','pedotmayer','pedotp',
              'pedotpcbm','pedottbapf6'],
    'electrolyte':['i2', 'liclo4spe', 'litfsi', 'speX2', 'lilco4',
                 'lino3', 'tabpf6', 'speX1', 'liclo4ecspe',
                 'ecspe', 'liclo4'],
    'anode':['ito', 'pfb', 'pfbirg', 'carbon']
                }
device_properties = {
    'name':[],
    'date':[],
    'cathode':[],
    'electrolyte':[],
    'anode':[]
}
for name in dnames:
    # If the entry has the anode, date, electrolyte, and cathode information that
    # we need, lets store it in device_properties
    bins = name.split('_');
    this_date = re.sub("[^0-9]", "", bins[0])
    this_cathode = list(set(bins) & set(film_materials['cathode']))
    this_electrolyte = list(set(bins) & set(film_materials['electrolyte']))
    this_anode = list(set(bins) & set(film_materials['anode']))
    if this_cathode and this_anode and this_electrolyte and this_date:
        device_properties['name'].append(name)
        device_properties['date'].append(this_date)
        device_properties['cathode'].append(this_cathode[-1])
        device_properties['electrolyte'].append(this_electrolyte[-1])
        device_properties['anode'].append(this_anode[-1])
```

Now we have a much cleaner dataframe to work with.

```python
cleandf = pd.DataFrame.from_dict(device_properties)
cleandf.head()
```

```python
#make a new column that counts how long a device was on the lifetime tester for
counts = col.Counter(cleandf.name)
counts = col.OrderedDict(sorted(counts.items()))
cleandf = cleandf.drop_duplicates(subset='name');
cleandf = cleandf.sort_index(by='name', ascending = True);
cleandf['weeks alive'] = pd.Series([x / 3 for x in counts.values()], index=cleandf.index)
```

```python
dcathode = pd.concat([ cleandf['cathode'], cleandf['weeks alive']],
                     axis = 1, keys = [ 'cathode', 'weeks alive'], )
danode = pd.concat([cleandf['anode'], cleandf['weeks alive']], axis = 1,
                   keys = ['anode', 'weeks alive'])
delectrolyte = pd.concat([cleandf['electrolyte'], cleandf['weeks alive']],
                         axis = 1, keys = ['electrolyte', 'weeks alive'])
datedf = pd.concat([cleandf['date'], cleandf['weeks alive']],
                   axis = 1, keys = ['date', 'weeks alive'])
```

```python
datedf.boxplot(column = 'weeks alive', by='date', rot='90')
```
